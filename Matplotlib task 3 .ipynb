# 10 Plot a pie chart using plt.pie()

```python
labels = ['Apples', 'Bananas', 'Cherries']
sizes = [30, 40, 30]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

# 11.Use subplots to display multiple plots



```python
fig, axs = plt.subplots(2, 1)
axs[0].plot([1, 2, 3], [4, 5, 6])
axs[1].bar(['A', 'B', 'C'], [1, 2, 3])
plt.show()
```

# 12.Customize ticks and grid lines

```python
plt.plot([1, 2, 3], [2, 4, 1])
plt.xticks([1, 2, 3])
plt.yticks([0, 2, 4])
plt.grid(True)
plt.show()
```

# 15.Add annotations to plot points



```python
x = [1, 2, 3]
y = [4, 1, 3]
plt.plot(x, y, marker='o')
for i in range(len(x)):
    plt.annotate(f"({x[i]},{y[i]})", (x[i], y[i]))
plt.show()
```

# 16.Create a stacked bar chart



```python

labels = ['A', 'B', 'C']
x = np.arange(len(labels))
a = [3, 2, 1]
b = [2, 3, 4]
plt.bar(x, a, label='A')
plt.bar(x, b, bottom=a, label='B')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

# 17.Change background and axis colors

```python
fig, ax = plt.subplots()
ax.set_facecolor('lightyellow')
fig.patch.set_facecolor('lightblue')
ax.plot([1, 2, 3], [4, 1, 2])
plt.show()
```
